### 1. Load the data from the netCDF file using `xarray`

Let's go through the steps of opening our sea surface temperature dataset using `xarray` nd then dealing with missing data

```python
# open our file into a DataSet and take a look:
ds = xr.open_dataset('HadISST_sst.nc') 
ds
```

Because we'll only be using the SST data, we can pull out that variable into a DataArray to simplify things a bit:

```python
sst = ds.sst
sst
```

### Filling negative numbers

Recall from our earlier lecture that there are missing data in the netCDF that are not handled by the import command above.  Specifically, there are non-NaN numerical indicators of missing SST data in the presense of sea ice.  The next line handles this by using the `.where` syntax.  This line essentially says 'Hey, xarray, tell me where SSTs are greater than or equal to zero, and where they are not, make the values NaN':

```python
# turn negative SST values in NaN
sst = sst.where(sst >= 0, np.nan)
```

### 2. Dealing with longitude coordinate systems

Here we map the -180 to 180 system to the 0 to 360 system and assign that to the coordinates of SSt.  Because these will then be out of order, we're going to sort everything in sst by the longitude values (which now go from 0 to 360).  

```python
# change the longitudes from -180,180 to 0,360
sst = sst.assign_coords(longitude=(sst.longitude % 360))

# now, sort the coordinates and associated data so they are in the correct order from 0 to 360
sst = sst.sortby(sst.longitude)

# This is an alternative way to do the same thing, but I find it less intuitive compared to the single step above
# sst = sst.roll(longitude=int(len(sst['longitude']) / 2), roll_coords=True)
```

### Sanity check

A term of art in programming, but we'd like to make sure that after all the things we did to our data above, that things still look right - continents are in the right place, tropics are warm and poles are cold, etc:

```python
sst.mean(dim="time").plot() # this says 'hey, xarray, take the mean over all the time points and plot the 2d result
```

### 2. Extract a subset of data from the Pacific

```python
# latitude bounds
min_lat = -20.0
max_lat = 60.00

# longitude bounds
min_lon = 120
max_lon = 260

# select a slice and overwrite sst with the output and then take a look
sst = sst.sel(longitude=slice(min_lon,max_lon),latitude=slice(max_lat,min_lat))
sst
```

I'll also pull out the latitude and longitude coordiates from the DataArray for simplicity later:

```python
# extract coordinates from the DataArray
lat = sst.latitude
lon = sst.longitude
```

Let's do one more sanity check now on our DataArray.  If we've done what we think we did, the `.plot` should show mean SSTs in the tropical and north Pacific, with the warmest SSTs in the tropics, colder conditions at the poles, and the continents where they belong:

```python
sst.mean(dim="time").plot()
```

Look good? Let's more on. 

### 3. Create a 5-month seasonal mean from November through March

Here we calculate the November through March average SST value for each location in our domain. 

```python
# Calculate the 5-month rolling mean, skipping NaN values
sst_ = sst.rolling(time=5).mean(skipna=True)

# Select out month 3 (March) to get the average over NDJFM
winter = sst_[sst_.time.dt.month == 3]

# take a look
winter
```

### Removing a spurious year from our DataArray

Now, remember what we talked about in lecture.  Xarray in trying to be helpful calculated a 5-month average from November to March that includes November 1869 and December 1869, which don't actually exist in our dataset.  This means that our first value is not valid, and indeed is full of NaN values - we can see this with a simple plot:

```python
winter.isel(time=0).plot() # map is empty, compare to winter.isel(time=2).plot()
```

No problem, we can use xarray's `.drop_isel` to remove the first time dimension in the dataset.  

```python
# see if all the values in timestep 0 are null - they are (returns True)
winter.isel(time=0).isnull().all()

# look at winter again - note the 154 years of winter averages and the first date in March 1870
winter
```

Here we do the dropping of that first year:

```python
# use drop_isel to drop the first year and reassign to winter
winter = winter.drop_isel(time=0)

# look at winter again - note the 153 years of winter averages and the first date in March 1871
winter
```

### 4., 5., 6. Reshaping data and finding locations without missing data and removing the time-mean

Currently, we expect our covariance and SVD analysis to use a 2-dimensional matrix, but we have 3-dimensions.  We'll first `.reshape` the data:  

```python
# get the values from the winter DataArray - at this point they are still 3-dimensional
sst2d = winter.values

# record their original dimensions before we reshape
ntime, nrow, ncol = sst2d.shape

# now reshape, we want rows of the time variable and columns of location - we use 'order=F' because this gives us the correct reshaping
sst2d = np.reshape(sst2d, (ntime, nrow*ncol), order='F')

# here we say 'tell me where there is NOT even a single NaN in the column.  Returns an array with the size of the columns of sst2d
noMissing = ~np.isnan(sst2d).any(axis=0)

# create a new array with only the columns of sst2d that are not missing any data 
sst2dcomplete = sst2d[:, noMissing]

# check - are there any NaN values?  count_nonzero function (recursively) counts how many elements have nonzero or a Boolean method evaluated to True.
print(np.count_nonzero(np.isnan(sst2dcomplete))) # should be zero, indicating the np.isnan contains no True booleans

# calculate the anomalies from the long-term seasonal mean for each column
sst_anomalies = sst2dcomplete - sst2dcomplete.mean(axis=0)
sst_anomalies
```

### 7. Calculating a covariance matrix

```python
# There are two ways to calculate the covariance matrix - both should work
C = np.cov(sst2dcomplete,rowvar=False,ddof=1)
```

### 8. The Singular Value Decomposition

We now move forward and do our Singular Value Decomposition.  Because out matrix is still quitel large (9329 columns) we can use sp.sparse.linalg.svds to calculate the 'economy size' SVD for a small number of modes. 

```python
# nunber of modes to extract using the economy-sized SVD 
number_of_modes = 2

# Run the SVD!
# U, S, V = np.linalg.svd(sst2dcomplete,full_matrices=False) # running this on the covariance matrix would take forever
U, S, V = sp.sparse.linalg.svds(C,k=number_of_modes) # economy sized, only get the first two eigenvalues/eigenvectors

# SciPy does not guarantee the eigenmodes are in order from most to least variance, so we have to do it
sort_index = np.argsort(S)[::-1] # sort in descending order from high to low
S = S[sort_index] # apply the index
U = U[:,sort_index] # apply the index

## that the eigenvectors in U and put them into an array for plotting
# create an empty matrix full of NaNs with the same number of columns as the original 2-dimensional data and 2 columns
eof = np.ones([nrow*ncol,number_of_modes]) * np.NaN
eof[noMissing,:] = U # take the eigenvalue and put them back in the right place in the eof array 
eof = eof.reshape([nrow,ncol,number_of_modes], order='F') # reshape back to the original dimensions of latitude and longitude plus the eof dimension
```

### 9. Calculating the 'PCs' - the time component

```python
# PCs come from the projection of the eigenvectors on the original centered data matrix
pcs = sst_anomalies @ U
```

### 10. Calculate the variance explained by each eigen-mode

```python
# fraction of variance accounted for by each mode is the eigenvalues divided by the total variance from the dataset
variance_explained = S/np.sum(np.var(sst2dcomplete,axis=0,ddof=1))
variance_explained*100 # gives percent variance explained
```

### 11. and 12. Plotting EOF patterns and PC Scores

```python
fig=plt.figure(layout="constrained")
gs = fig.add_gridspec(2, 2) # 2 row and 3 column spaces
gs.update(wspace=-1, hspace=-0)

projection=ccrs.Miller(central_longitude=260)# centered on North America

# specifying the levels gives us more control on the map
levels = np.linspace(-0.035, 0.035, 31)

ax1 = fig.add_subplot(gs[0, 0], projection=projection) # map plot in 1st column of grid
fx1 = ax1.contourf(lon,lat,eof[:,:,0],levels=levels, cmap='bwr',add_colorbar=False, transform=ccrs.PlateCarree())
ax1.set_extent([120, 260, -20, 61],crs=ccrs.PlateCarree())
ax1.coastlines(color="black")  # we can pass arguments to .coastlines function
ax1.add_feature(cfeature.LAND, color='black')
ax1.set_title('EOF1 Spatial Pattern ('+str(round(variance_explained[0]*100))+'% variance)', fontsize=12)
cb1 = plt.colorbar(fx1, orientation='vertical',location='left',shrink=0.6,ticks=[-0.03, -0.01, 0.01, 0.03])
cb1.set_label('loadings', fontsize=10) 

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(winter.time,pcs[:,0],color="black")
ax2.set_title('PC1 Time Series', fontsize=12)
ax2.tick_params(axis="x",labelsize=9)
ax2.tick_params(axis="y",labelsize=9)

ax3 = fig.add_subplot(gs[1, 0], projection=projection) # map plot in 1st column of grid
fx3 = ax3.contourf(lon,lat,-eof[:,:,1],levels=levels, cmap='bwr',add_colorbar=False, transform=ccrs.PlateCarree())
ax3.set_extent([120, 260, -20, 61],crs=ccrs.PlateCarree())
ax3.coastlines(color="black")  # we can pass arguments to .coastlines function
ax3.add_feature(cfeature.LAND, color='black')
ax3.set_title('EOF2 Spatial Pattern ('+str(round(variance_explained[1]*100))+'% variance)', fontsize=12)
cb3 = plt.colorbar(fx3, orientation='vertical',location='left',shrink=0.6,ticks=[-0.03, -0.01, 0.01, 0.03])
cb3.set_label('loadings', fontsize=10) 

ax4 = fig.add_subplot(gs[1, 1])
ax4.plot(winter.time,-pcs[:,1],color="black")
ax4.set_xlabel('Year',fontsize=12)
ax4.set_title('PC2 Time Series', fontsize=12)
ax4.tick_params(axis="x",labelsize=9)
ax4.tick_params(axis="y",labelsize=9)
```

### 13. How does your map of loadings compare to that in Deser and Blackmon (1995)? What might explain any differences you observe?

Here is the corresponding figure from Deser and Blackmon (1995):

![Alt text](image.png)

The pattern shown for EOF1 is similar - positive loadings in the eastern tropical Pacific and extending along the coast of the Americas both north and south and then a horseshoe of negative loadings in the  northwestern extra-tropical Pacific and the equatorial western Pacific.  The magnitude of these loadings is different, because Deser and Blackmon use the correlation between PC1 and the original data, so the values are normalized.  EOF2 has some similar patterns - loadings are weakly negative int he eastern tropical equatorial Pacific and strongly positive in the northwestern Pacific.  However, the spatial structure is different - our negative loadings are more confined to the equator than in Deser and Blackmon, and the northwest Pacific pattern is less prominent and stronger near the coast.  Once again, the magnitude of these loadings is different, because Deser and Blackmon use the correlation between PC2 and the original data for the map.  Possible reasons for the differences include: (1) a different time period over which the EOFs are calculated, the different datasets (COADS vs HasISST) and spatial resolution (COADS is 4x4!), and a different way of expressing the loadings (using correlation maps instead of the eigenvector values).
