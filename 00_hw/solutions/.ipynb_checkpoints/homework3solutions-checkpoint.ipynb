### 1. Import packages

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

# Cartopy gives a lot of warnings when downloading map data, which we'll suppress
import warnings
warnings.filterwarnings("ignore") 

# you can omit the line below if you'd like, but I really don't like the default fonts in Python, so I switch to Helvetica
plt.rcParams['font.family'] = 'Helvetica'
```

### 2. Read in the cereal production data (in metric tons) from the Food and Agriculture Organization of the United Nations

```python
df = pd.read_excel('cereal_production_WB_FAO_2022.xlsx',index_col=[0]) # the second input says to make the 0th column into the index column - the years
df
```

### 3. Identify countries (in the DataFrame columns) with any missing data points and then create a new DataFrame with only the data from those countries with complete data series over the complete time period

```python
# identify columns where there are no null (NaN) values 
all_available = df.isnull().sum()==0

# pull out countries without missing data and put them in a new DataFrame
cereals = df.loc[:,all_available].copy()

# get years a values in a list to use for plotting
years = df.index.values.tolist()
```

### 4. Modify each column so that it has a mean of zero and a standard deviation of one. 

Here I'm using the Scipy `zscore` method to do this, but you could also do it by subtracting the mean and dividing by the standard deviation.

```python
cereals = sp.stats.zscore(cereals)
```

### 5. Calculate the covariance of the normalized cereal data, calculate the SVD, and then derive the time signals

```python
# covairnace matrix
C = np.cov(cereals,rowvar=False,ddof=1)

# SVD
eofs,evals,V = np.linalg.svd(C)

# check the signs of the eigenvector values for EOF1 - we may want to flip the arbitrary sign to make one or more of the components more easily interpretable
plt.hist(eofs[:,0],label='original sign')

# flip the sign of EOF1 so that they are mostly positive, for interpretive reasons
eofs[:,0] = -eofs[:,0]

# recheck signs
plt.hist(eofs[:,0],label='flipped sign')
plt.legend()

# then derive the PCs by projecting the eigenvectors back on the original data
pcs = cereals @ eofs
```

Let's plot a quick sanity check to make sure the PCs look reasonable:

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)
pcs.plot(y=[0],ax=ax1)
ax1.legend(['MODE 1'])
ax1.set_xlabel("")

pcs.plot(y=1,ax=ax2)
ax2.legend(['MODE 2'])
ax2.set_xlabel('YEAR')
plt.tight_layout()
```

### 5. Scale the PCs by the square root of their eigenvalues

```python
scaled_pcs = pcs/np.sqrt(evals)
```

Let's do another quick sanity check - note how the scaling affect the PCs units (or we might say their magnitude)

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)
ax1.plot(years,scaled_pcs[0])
ax1.legend(['MODE 1'])
ax1.set_xlabel("")

ax2.plot(years,scaled_pcs[1])
ax2.legend(['MODE 2'])
ax2.set_xlabel('YEAR')
plt.tight_layout()
```

### 6. Identify which countries – and therefore, which columns of your DataFrame – are contributing to your two primary signals.

```python
# get the list of countries
countries_names = cereals.columns.values
ncountries = len(countries_names) #145

# set a eienvector loading threshold for the contributors to EOF1
eof1_contributor_threshold = 0.0

# create an index for those contributors
eof1_contributors = eofs[:,0]>eof1_contributor_threshold

# use that as an index into the list of countries we extracted
print(np.count_nonzero(eof1_contributors)) # 124 countries
countries_names[eof1_contributors]
```

```python
# what do the eigenvector loadings look like on EOF2? 
plt.hist(eofs[:,1])

# set a threshold based on the histogram
eof2_contributor_threshold = 0.10

# create an index for those contributors
eof2_contributors = eofs[:,1]>eof2_contributor_threshold

# use that as an index into the list of countries we extracted
print(np.count_nonzero(eof2_contributors)) # 9 countries
countries_names[eof2_contributors]
```

### 8. Plot the scaled PC signals for EOF1 and EOF2 on the same plot, overlaying the individual country lines

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)

cereals.loc[:,eof1_contributors].plot(ax=ax1,linewidth=0.5)
scaled_pcs.plot(y=[0],ax=ax1,color="black",linewidth=3)
ax1.legend("")
ax1.set_xlabel("")
ax1.set_ylabel('NORMALIZED VALUES')


cereals.loc[:,eof2_contributors].plot(ax=ax2,linewidth=0.5)
scaled_pcs.plot(y=[1],ax=ax2,color="black",linewidth=3)
ax2.legend("")
ax2.set_xlabel('YEAR')
ax2.set_ylabel('NORMALIZED VALUES')

plt.tight_layout()
```

### 9. Isolate and plot the original data associated with the negative (assuming the majority of your EOF1 loadings are positive!) loadings in eigenvector #1

```python
# set a eienvector loading threshold for the countries with a negative loading on EOF1
eof1_negative_threshold = 0.0

# create an index for those contributors
eof1_negative_loading = eofs[:,0]<eof1_negative_threshold
```

```python
fig, (ax1) = plt.subplots(nrows=1,ncols=1)

cereals.loc[:,eof1_negative_loading].plot(ax=ax1,linewidth=0.5)
scaled_pcs.plot(y=[0],ax=ax1,color="black",linewidth=3)
ax1.legend("")
ax1.set_xlabel("")
ax1.set_ylabel('NORMALIZED VALUES')
ax1.set_xlabel("YEAR")
```

It is clear that the countries with a negative loading on EOF2 did so because they have a downward trend in cereals production through time, the opposite of the dominant positive loading countries with an overall upward trend in cereals production.

### 11. Calculate the variance explained by each eigen-mode 

```python
variance_explained = 100 * (evals/np.sum(evals))
```

### 12. Let’s create what is known as a scree plot

```python
fig, (ax1) = plt.subplots(nrows=1,ncols=1)

ax1.stem(np.arange(1,len(evals)+1),variance_explained,basefmt=" ")
ax1.set_xlabel('Eigenvalue Rank')
ax1.set_ylabel("Percent Variance Explained")
ax1.set_xlim((0,11))
ax1.set_ylim((0,70))
```

### 13. Gaussian Significance Test (Rule N)

```python
# how many iterations? Be sure to start small until you know how long this might take
iterations = 1000

# save the size of our data matrix
mrows, ncols = cereals.shape

# create an array full of NaN values to acccept the random eigenvalues we'll generate 
evals_simulated = np.full((iterations,len(evals)),np.nan) # rows of iteration number, columns of the length of the eigenvalue vector
evals_simulated_var = np.full((iterations,len(evals)),np.nan)

# loop through by iteration ... 
for i in range(iterations):
    x = np.random.normal(cereals.values.mean(axis=0),cereals.values.std(axis=0),cereals.values.shape) # random normal matrix the size of our data with mean and standard deviation of our data
    x = x - x.mean(axis=0) # remove the mean
    Cx = np.cov(x,rowvar=False,ddof=1) # get the covariance of the random matrix
    Ux,Sx,Vx = np.linalg.svd(Cx) # factor the covariance matrix into eigenvalues and eigenvectors
    evals_simulated[i,:] = Sx # save the simulated eigenvalues
    evals_simulated_var[i,:] = 100 * (Sx/np.sum(Sx)) # could save some time by moving this operation out of the loop

ci90zvar = np.percentile(evals_simulated_var,90,axis=0) #
ci90zvar = np.percentile(evals_simulated_var,5,axis=0) #
```

Now let's plot our variance explained in a stem (scree) plot along with the 95% confidence interval:

```python
fig, (ax1) = plt.subplots(nrows=1,ncols=1,layout="constrained")

ax1.stem(np.arange(1,len(evals)+1),variance_explained,basefmt=" ",label="MODE VARIANCE EXPLAINED")
ax1.set_xlabel('Eigenvalue Rank')
ax1.set_ylabel("Percent Variance Explained")
ax1.set_xlim((0,11))
ax1.set_ylim((0,70))
ax1.plot(np.arange(1,len(evals)+1),ci90zvar,linestyle="--",color="red",label="90% CI")

ax1.legend()
```
